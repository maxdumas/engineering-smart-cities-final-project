We are mostly interested in highrises, so let's just focus on finding a representative building height that would require a water tower in Manhattan.

So let's find the heights of all of the buildings in Manhattan that are over 6 stories. A 6 story building is about 60 to 75 feet tallâ€”let's just say 60 feet.

We know that a BIN that start with 1 indicates a building that is in Manhattan. 

```python
df[df.HEIGHTROOF > 60]
```

```python
mnh_buildings_with_towers_df = df.query('BIN.astype("str").str.startswith("1") & HEIGHTROOF > 60')
```

```python
mnh_buildings_with_towers_df
```

```python
mean_building_height = mnh_buildings_with_towers_df.HEIGHTROOF.mean()
mean_building_height
```

```python
mean_building_area = mnh_buildings_with_towers_df.geometry.area.mean()
mean_building_area
```

```python
BUILDING_STORY_HEIGHT_FT = 10
mean_building_stories = mean_building_height / 10
mean_building_stories
```

```python
usable_building_stories = mean_building_stories - 1
mean_livable_sq_ft_per_building = usable_building_stories * mean_building_area
mean_livable_sq_ft_per_building
```

```python
SQ_FT_PER_PERSON = 300

mean_people_per_building = mean_livable_sq_ft_per_building / SQ_FT_PER_PERSON
mean_people_per_building
```

```python
PER_PERSON_MEAN_WATER_CONSUMPTION_LITER_PER_DAY = 420.18

mean_building_water_consumption_per_day = PER_PERSON_MEAN_WATER_CONSUMPTION_LITER_PER_DAY * mean_people_per_building
mean_building_water_consumption_per_day
```

```python
# We compute the tank outflow per minute as a result of the constant draw from usage.
tank_outflow_l_per_minute = mean_building_water_consumption_per_day / 24 / 60
tank_outflow_l_per_minute
```

```python
MEAN_WATER_TANK_HEIGHT_M = 3.6576
MEAN_WATER_TANK_DIAMETER_M = 3.6576
MEAN_WATER_TANK_RADIUS_M = MEAN_WATER_TANK_DIAMETER_M / 2.0
L_PER_CUBIC_M = 1_000

mean_water_tank_volume_l = np.pi * MEAN_WATER_TANK_RADIUS_M ** 2 * MEAN_WATER_TANK_HEIGHT_M * L_PER_CUBIC_M
mean_water_tank_volume_l
```

```python
# TODO: We can't actually let the tank empty, both to maintain water pressure
# and to adhere to fire codes. So in reality the tank needs to be refilled more
# often than this. Should estimate that somehow.

n_tank_cycles_per_day = mean_building_water_consumption / mean_water_tank_volume_l
n_tank_cycles_per_day
```

```python
hours_per_tank_cycle = 24 / n_tank_cycles_per_day
hours_per_tank_cycle
```

```python
# The amount of time it takes to completely refill the tank from empty. Sources
# indicate that this takes from 2â€“3 hours.
TANK_FILLING_TIME_MINUTES = 180.0
tank_filling_rate_l_per_minute = mean_water_tank_volume_l / TANK_FILLING_TIME_MINUTES
tank_filling_rate_l_per_minute
```

```python
excess_pump_capacity_l_per_minute = tank_filling_rate_l_per_minute - tank_outflow_l_per_minute
excess_pump_capacity_l_per_minute
```

```python
excess_pump_capacity = excess_pump_capacity_l_per_minute / tank_filling_rate_l_per_minute
excess_pump_capacity
```

```python
tank_filling_time_with_outflow_hours = TANK_FILLING_TIME_MINUTES * (2 - excess_pump_capacity) / 60
tank_filling_time_with_outflow_hours
```

```python
total_filling_discharge_period_hours = tank_filling_time_with_outflow_hours + hours_per_tank_cycle
total_filling_discharge_period_hours
```

Ok, so we now know that, for the mean building in Manhattan that requires a
water tank, it has a pumping capacity that exceeds the mean outflow as a result
of water usage by about 65%. That means we can easily outpace the demand and
increase the water level of the tank even as the building is consuming its
average amount of water, though it will slow the ability of the tank to refill.
Our goal is to use the pump as little as possible, so we want to essentially
only run the pump full blast when power is cheapest and then otherwise keep it
completely off.

According to the above, we'd basically have to spend 4 hours filling the tank,
and then we'd get about 8.6 hours for the tank to draw down before we have to
fill it again, just off of normal water usage. This gives us some leeway in our
ability to draw additional water and in our ability to time the pumping and
additional discharge.

Let's think about how increasing water demand would affect this capacity.

Essentially what I'm implying above is that we can afford to discharge about 138
l/m more of water in the mean building before the water tank will on average be
unable to refill faster than it is draining. At 138 l/m of average discharge,
the pump would be required to run constantly to maintain the water level, which
would mean that there would be no benefit to this process as there would be no
ability to advantageously time the pumping or discharging to flatten loads. We'd
just be wasting a bunch of energy.

Meanwhile, if we run the pump only exactly as much as needed to accommodate the
usage of the humans in the building, we of course hit maximal efficiency in that
we're only discharging (and pumping) the water that we need. We can't control
when people are discharging the water, but we can control when we pump and when
we choose to add additional discharge. So the key is to find where we have
leeway in that curve to forecase user demand patterns and to choose when we pump
and when we additionally dischage so as to flatten the demand curve as much as
possible while minimized water waste and ensuring the residents have enough
water.
