```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from scipy.special import expit

from esc.util import MEAN_WATER_TANK_HEIGHT_M
from esc.reward import reward_high_tank_head, reward_low_energy_cost
```

```python
tank_heads = np.linspace(0, MEAN_WATER_TANK_HEIGHT_M, 100)
total_energy_cost = np.linspace(0, 200, 200)
```

```python
plt.plot(tank_heads, reward_high_tank_head(tank_heads))
plt.axvline(MEAN_WATER_TANK_HEIGHT_M, color="r")
plt.grid()
```

```python
plt.plot(total_energy_cost, reward_low_energy_cost(total_energy_cost))
plt.grid()
```

```python
reward_low_energy_cost(0)
```

```python
X, Y = np.meshgrid(total_energy_cost, tank_heads)
```

```python
Z = reward_high_tank_head(Y) * reward_low_energy_cost(X)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
fig.set_figwidth(12)
fig.set_figheight(12)
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
ax.set_xlabel("Total Energy Cost")
ax.set_ylabel("Tank Head")
ax.set_zlabel("Reward")
```
